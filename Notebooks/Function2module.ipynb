```python
from pathlib import Path

def TitleInfo(currentFile):
    Title=[]
    with open(currentFile) as fileIn:
        print(':\033[1;31mI\033[1;m')
        print('PDB File:\033[1;31m %s\033[1;m'%currentFile)
        line = fileIn.readline()
        while line:
            if line.startswith('TITLE'):
                Title.append(line)

            line = fileIn.readline()
        if len(Title) == 1:
            Str = "".join(Title)
            x = Str.replace('TITLE', '')
            Str1 = x.lstrip()
            print('Title: %s'%Str1)
        if len(Title) > 1:
            #Str = "".join(l)
            t =(Title[0])
            z = (Title[1])
            t1 = t.replace('TITLE', '')
            z1 = z.replace('TITLE', '')
            z2 = z1.replace('2', '')
            t2 = t1.strip()
            z3 = z2.strip()
            print('Title:%s'%t2+z3)
            #return Title

        

def ReturnChain(currentFile):
    with open(currentFile) as fileIn:
            lines = fileIn.readlines()
            Chainlist=[]
            for line in lines:
                if line.startswith('SEQRES'):
                    List = line.split()
                    Chainlist.append(List[2])
            #print(Chainlist)
            Chain = set(Chainlist)
            chain = sorted(list(Chain))
            return chain

    
def seqprint(seqList): # prints 50 a/a per line
    List = seqList
    seq = [aaCode[b] for b in List]
    seq2 =" ".join(seq)
    seq3 = seq2.replace(" ","")
    read_length = len(seq3)
    #print(len(seq3))
    if read_length < 50:
        print(seq3)
    else:
        char_per_line = 50
        for base in range(0,len(seq3),char_per_line):
            print (seq3[base:base+char_per_line])
            
                
def AminoAcidCount(currentFile): #final, prints No. of amino acid,helix and sheet info. by call seqprint function
    Chains = ReturnChain(currentFile)
    chainlist = Chains
    chainlist.insert(len(chainlist)-1,'and')
    print('CHAINS:',' '.join(chainlist))
    with open(currentFile) as fileIn:
            line = fileIn.readline()
            A=B=C=D = 0
            sequenceA=[]
            sequenceB= []
            sequenceC=[]
            sequenceD= []
            helix_countA = 0
            helix_countB = 0
            helix_countC = 0
            helix_countD = 0
            sheet_countA = 0
            sheet_countB = 0
            sheet_countC = 0
            sheet_countD = 0
            while line:
                if line.startswith('SEQRES'):
                    words = line.split()
                    chaintype = words[2]
                    if line.startswith('SEQRES') and chaintype == 'A':
                        letter = line.split()
                        #A = letter[3]
                        for word in letter:
                            if word in aminoacidsCode.values():
                                sequenceA.append(word)

                    if line.startswith('SEQRES') and chaintype == 'B':
                        letterB = line.split()
                        #B = letterB[3]
                        for word in letterB:
                            if word in aminoacidsCode.values():
                                sequenceB.append(word)

                    if line.startswith('SEQRES') and chaintype == 'C':
                        letterC = line.split()
                        #C = letterC[3]
                        for word in letter:
                            if word in aminoacidsCode.values():
                                sequenceC.append(word)

                    if line.startswith('SEQRES') and chaintype == 'D':
                        letterD = line.split()
                        #D = letterD[3]
                        for word in letter:
                            if word in aminoacidsCode.values():
                                sequenceD.append(word)

                    else:
                        pass
                        
                        
                if line.startswith('HELIX'):
                        helixInfo = line.split()
                        chaintype = helixInfo[4]
                        for chain in Chains:
                            if chain == chaintype and chaintype == 'A':
                                helix_countA = helix_countA + 1
                            elif chain == chaintype and chaintype == 'B':
                                helix_countB = helix_countB + 1
                            elif chain == chaintype and chaintype == 'C':
                                helix_countC = 1
                            elif chain == chaintype and chaintype == 'D':
                                helix_countD = 1
                            else:
                                pass
                            
                if line.startswith('SHEET'):
                        sheetInfo = line.split()
                        chaintype = sheetInfo[5]
                        for chain in Chains:
                            if chain == chaintype and chaintype == 'A':
                                sheet_countA = sheet_countA + 1
                            elif chain == chaintype and chaintype == 'B':
                                sheet_countB = sheet_countB + 1
                            elif chain == chaintype and chaintype == 'C':
                                sheet_countC = sheet_countC + 1
                            elif chain == chaintype and chaintype == 'D':
                                sheet_countD = sheet_countD + 1
                            else:
                                pass
                     
                line = fileIn.readline()
            
            if 'A' in Chains:
                print('- Chain A')
                print('Number of amino acids: %s'%len(sequenceA))
                print('Number of helix: %s'%helix_countA)
                print('Number of sheet: %s'%sheet_countA)
                seqprint(sequenceA)
                
            if 'B' in Chains:
                print('- Chain B')
                print('Number of amino acids: %s'%len(sequenceB))
                print('Number of helix: %s'%helix_countB)
                print('Number of sheet: %s'%sheet_countB)
                seqprint(sequenceB) 
            if 'C' in Chains:
                print('- Chain C')
                print('Number of amino acids: %s'%len(sequenceC))
                print('Number of helix: %s'%helix_countC)
                print('Number of sheet: %s'%sheet_countC)
                seqprint(sequenceC)
            if 'D' in Chains:
                print('- Chain D')
                print('Number of amino acids: %s'%len(sequenceD))
                print('Number of helix: %s'%helix_countD)
                print('Number of sheet: %s'%sheet_countD)
                seqprint(sequenceD)
                                            
aminoacidsCode= {1:'ALA', 2: 'CYS',3: 'ASP',4:'GLU',5:'PHE',6:'GLY',7:'HIS',8:'ILE',9:'LYS',10:'LEU',11: 'MET',
             12:'ASN',13:'PRO',14: 'GLN',15:'ARG',16:'SER',17:'THR',18:'VAL',19:'TRP',20:'TYR'}

aaCode= {'ALA':'A','CYS':'C','ASP':'D','GLU':'E','PHE':'F','GLY':'G','HIS':'H','ILE':'I','LYS':'K','LEU':'L','MET':'M',
             'ASN':'N','PRO':'P','GLN':'Q','ARG':'R','SER':'S','THR':'T','VAL':'V','TRP':'W','TYR':'Y'}

def main():
    filename = Path(input('enter path:'))


    if  filename.exists():
        #print('file successfuly loaded')
        currentFile = filename.name
        TitleInfo(currentFile)
        AminoAcidCount(currentFile)
        
    else:
        print('Enter a Valid Path for a PDB file: %s'%filename.name)

main()
```
