```python
def FileUpdate(currentFile):
    print('Do you want to replace the current file %s ?'%currentFile)
    selection = input("Please enter 'yes' to replace the file or 'no' to proceed: ")
    if selection in ('yes','no'):
        if selection.lower() == 'yes':
            print(selection.lower())
            filename = Path(input('enter path to enter the new file:'))
            if  filename.exists():
                print('New file successfuly loaded')
                File2 = filename.name
                if File2 == currentFile:
                    print('The file %s already loaded'%File2)
                else:
                    currentFile = File2
                    print('* The new Current PDB:\033[1;31m %s \033[1;m'%currentFile)
                    funct1.valid_file(currentFile)
                    
            else:
                print('Enter a Valid Path for a PDB file: %s'%filename.name)
        else:
            selection.lower() == 'no'
            print(selection.lower())
            pass
                
                
```
